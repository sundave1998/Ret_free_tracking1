```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
retraction_free_time = []

retraction_time = []

dimension = []

dimension.append(1000)
retraction_free_time.append(123.27846169471741)
retraction_time.append(211.65206575393677)

dimension.append(500)
retraction_free_time.append(34.137083768844604)
retraction_time.append(53.4998300075531)

dimension.append(300)
retraction_free_time.append(7.90181827545166)
retraction_time.append(12.799596548080444)


dimension.append(100)
retraction_free_time.append(3.2091784477233887)
retraction_time.append(3.590099811553955)


dimension.append(50)
retraction_free_time.append(1.5651731491088867)
retraction_time.append(1.5984487533569336)
```

```python
# species = ("Adelie", "Chinstrap", "Gentoo")
label1 = "Retraction-free Tracking"
label2 = "Riemannian DGD [number]"

plt.rcParams["figure.figsize"] = (6,4)

# figure(figsize=(4,3))
x = np.arange(len(dimension))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

# for i in range(len(dimension)):
offset = width * multiplier
rects = ax.bar(x + offset, retraction_free_time, width, label=label1)
rects.datavalues = np.around(rects.datavalues,2)
ax.bar_label(rects, padding=3, fmt='%.2f')

multiplier += 1

offset = width * multiplier

rects = ax.bar(x + offset, retraction_time, width, label=label2)
rects.datavalues = np.around(rects.datavalues,2)
ax.bar_label(rects, padding=3, fmt='%.2f')
multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('CPU time (s)')
ax.set_xlabel('Problem Dimension')
ax.set_title('Elapsed Time for Convergence')
ax.set_xticks(x + width/2, dimension)
ax.legend(loc='upper right')
ax.set_ylim(0, 240)

# plt.show()
# plt.savefig("time_comp.pdf")
```
